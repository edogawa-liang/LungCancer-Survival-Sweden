# Data Preprocessing for Lung Cancer Dataset

```python
import pandas as pd
data = pd.read_csv('data/Lung Cancer.csv')
data
```

```python
sweden_data = data[data['country'] == 'Sweden'].drop(columns=['country'])
sweden_data
```

```python
# calculate the degree of censoring
print(sweden_data['survived'].sum())
print(33161-7165)
print(sweden_data['survived'].sum()/33161)
```

```python
print(sweden_data['smoking_status'].unique())
print(sweden_data['treatment_type'].unique())
```

```python
gender_map = {"Female": 0, "Male": 1}
family_history_map = {"No": 0, "Yes": 1}
smoking_map = {
    "Never Smoked": 0,
    "Passive Smoker": 1,
    "Former Smoker": 2,
    "Current Smoker": 3
}
cancer_stage_map = {
    "Stage I": 0,
    "Stage II": 1,
    "Stage III": 2,
    "Stage IV": 3
}
treatment_map = {
    "Radiation": 0,
    "Chemotherapy": 1,
    "Surgery": 2,
    "Combined": 3
}

sweden_data["gender"] = sweden_data["gender"].map(gender_map)
sweden_data["family_history"] = sweden_data["family_history"].map(family_history_map)
sweden_data["smoking_status"] = sweden_data["smoking_status"].map(smoking_map)
sweden_data["cancer_stage"] = sweden_data["cancer_stage"].map(cancer_stage_map)
sweden_data["treatment_type"] = sweden_data["treatment_type"].map(treatment_map)
```

```python
sweden_data.to_csv('data/Sweden_Lung_Cancer.csv', index=False)
```
